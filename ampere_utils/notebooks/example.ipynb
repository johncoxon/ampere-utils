```python
from ampere_utils.read import ncdf_file_from_path
from ampere_utils.plot import polar_plot
from matplotlib import pyplot as plt
```

# Read the data

Read the AMPERE data from some path on your computer, and then excerpt the various variables so that they can be passed to `polar_plot`.

```python
filepath = "/Volumes/Datasets/AMPERE/ncdf/grd/2011/ampere.20110101.0000.86400.120.north.grd.ncdf"
```

```python
ampere_data = ncdf_file_from_path(filepath)
mlt = ampere_data["mlt"].reshape(24, 50)[:, 0]
colat = ampere_data["colat"].reshape(24, 50)[0, :]
j = ampere_data["j"][100, :].reshape(24, 50)
hemisphere = ampere_data["hemisphere"]
```

# Plot the data

Plot the data using `polar_plot` and give it a colour bar.

```python
fig, ax = plt.subplots(1, subplot_kw={"polar": True})
mesh = polar_plot(mlt, colat, j, hemisphere, ax)
plt.colorbar(mesh)
```
